```python
def vocales(palabra):
  vocales = ["A","E","I","O","U","a","e","i","o","u"]
  contador = 0
  for chr in palabra:
    if chr in vocales:
      contador += 1
  return contador

palabra = input("Ingrese una palabra: ")

print(palabra," tiene ",vocales(palabra)," vocales")
```

```python
def cargar_personas():
  i = 0
  lista_nombres = []
  lista_edades = []
  while i < 5:
    lista_nombres.append(input("Ingrese un su nombre: "))
    lista_edades.append(int(input("Ingrese un su edad: ")))
    i += 1
  return lista_nombres,lista_edades

def imprimir_mayores(lista_nombres,lista_edades):
  print("═════════════","Mayores de 18","═════════════", sep = "\n")
  index = 0
  for edad in lista_edades:
    if edad >= 18:
      print(lista_nombres[index]," tiene ", edad)
    index += 1

def edad_promedio(lista_edades):
  acumulador = 0
  contador = 0
  for edad in lista_edades:
    acumulador += edad
    contador += 1
  return acumulador / contador

lista_nombres,lista_edades = cargar_personas()
imprimir_mayores(lista_nombres,lista_edades)

print("La edad promedio es ", edad_promedio(lista_edades))

```
